### Create a title: 

```python
data_file = './iris.csv'

def read_csv(filename):
    with open(filename, 'r') as f:
        data = list(csv.reader(f))
    return data

flowers = read_csv(data_file)
# print(flowers[0])

colors = {"Iris-setosa": "#2B5b84", "Iris-versicolor": "g", "Iris-virginica": "purple"}
flowers.pop()
# print(flowers)

for species, group in groupby(flowers, lambda i: i[4]):
    flower_cat = list(group)
    sepal_length = [float(flower[0]) for flower in flower_cat]
    sepal_width = [float(flower[1]) for flower in flower_cat]
    plt.scatter(sepal_length, sepal_width, s=25, c=colors[species], label=species)

plt.title("Flower Power Hour", fontsize=12)
plt.xlabel("Sepal Length (cm)", fontsize=12)
plt.ylabel("Sepal Width (cm)", fontsize=10)
plt.legend(loc="best").get_frame().set_color("red")

plt.show()
```

## Histogram Representation:

```python
data_file = './iris.csv'
plt.figure(figsize=(10.5, 7.25)) # This is where the change is

def read_csv(filename):
    with open(filename, 'r') as f:
        data = list(csv.reader(f))
    return data

flowers = read_csv(data_file)

v_petal_length = []

num_bins = 6

for petal in range(0, len(flowers) -1):
    if flowers[petal][4] == 'Iris-virginica':
        v_petal_length.append(float(flowers[petal][2]))

plt.hist(v_petal_length, num_bins, facecolor="red", alpha=0.50)
plt.title("Iris-Virginica Petal Length", fontsize=12)
plt.xlabel("Petal Length (cm)", fontsize=10)
plt.ylabel("Probablity", fontsize=10)
plt.show
```

## Box Plot 

```python
data_file = './iris.csv'
plt.figure(figsize=(10.5, 7.25)) # This is where the change is

fig = plt.figure()

def read_csv(filename):
    with open(filename, 'r') as f:
        data = list(csv.reader(f))
    return data

flowers = read_csv(data_file)
flowers.pop()

petal_length = []

# num_bins = 20

for species, group in groupby(flowers, lambda i: i[4]):
    petal_length.append([float(petal[2]) for petal in group])

plt.boxplot(petal_length)
plt.axis([0, 4, 0, 10])
plt.xticks([1, 2, 3], ['Setossa', 'Versicolor', 'Virginica']) # This is P important

plt.title("Flower Power Hour Petal Length", fontsize=12, loc="right")

fig.savefig('iris_petal_data.png')
```

## A Second, Much Tidier Histogram:

```python
data_file = './iris.csv'
plt.figure(figsize=(10.5, 7.25))
plt.style.use('classic')

def read_csv(filename):
    with open(filename, 'r') as f:
        data = list(csv.reader(f))
    return data

flowers = read_csv(data_file)
# print(flowers)

v_petal_length = []

num_bins = 20

for petal in range(0, len(flowers) - 1):
    if flowers[petal][4] == 'Iris-virginica':
        v_petal_length.append(float(flowers[petal][2]))
# print(v_petal_length)

plt.hist(v_petal_length, num_bins, facecolor="red", alpha=0.50, hatch='////\\\\')
plt.title("Iris-Virginica Petal Length", fontsize=12)
plt.xlabel("Petal Length (cm)", fontsize=10)
plt.ylabel("Probability", fontsize=10)
plt.show()
```
